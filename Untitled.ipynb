```python
import pandas as pd

# Load datasets
true_df = pd.read_csv("C:/Users/gouri/Downloads/true.csv.csv")
fake_df = pd.read_csv("C:/Users/gouri/Downloads/fake.csv.xls", engine='python')  # or rename and save as .csv

# Add labels
true_df['label'] = 1  # Real
fake_df['label'] = 0  # Fake

# Combine
df = pd.concat([true_df, fake_df], ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import re

# Simple text cleaner
def clean_text(text):
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.lower()

df['text'] = df['text'].apply(clean_text)  # adjust column name if needed

# Split data
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

# Vectorize
tfidf = TfidfVectorizer(stop_words='english', max_df=0.7)
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec = tfidf.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train_vec, y_train)
pred_lr = lr.predict(X_test_vec)
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred_lr))

# Naive Bayes
nb = MultinomialNB()
nb.fit(X_train_vec, y_train)
pred_nb = nb.predict(X_test_vec)
print("Naive Bayes Accuracy:", accuracy_score(y_test, pred_nb))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(lr, X_test_vec, y_test)
plt.title("Logistic Regression Confusion Matrix")
plt.show()

ConfusionMatrixDisplay.from_estimator(nb, X_test_vec, y_test)
plt.title("Naive Bayes Confusion Matrix")
plt.show()
```
